```python
import requests
import json
from pandas import json_normalize
import pandas as pd
from scrapethat import *
```

```python
# Creating function to get the data from OPEN FOOD FACTS API
def get_cat_data(category, num_products):
    url = "https://world.openfoodfacts.org/cgi/search.pl"

    # Defining parameters for the request on the "category" tag
    params = {
        'action': 'process',
        'tagtype_0': 'categories',
        'tag_contains_0': 'contains',
        'tag_0': category,
        'page_size': num_products,
        'json': 1,
    }

    # Making the request
    response = requests.get(url, params = params)
    
    # Checking if request successful against status code 200
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error with request: {response.status_code}")
        return None
```

```python
# Creating function to extract the product data from the JSON database
def create_df(off_data):
    # Extracting the product data
    products = off_data.get('products', [])

    # Creating the OPEN FOOD FACTS DataFrame from the product data
    off_df = pd.json_normalize(products)
    return off_df
    
```

```python
# Selecting data for one category (database is huge https://world.openfoodfacts.org/categories)
category = "Chocolate bars"
num_products = 200
```

```python
# Get Open Food Facts data
off_data = get_cat_data(category, num_products)

# Create DataFrame from the data with a limit on number of columns (over 3000)
off_df = create_df(off_data)
off_df = off_df.iloc[:, :50]

# Display the DataFrame
off_df
```

```python
# Creating function to get the links
def get_links(off_data):
    products = off_data.get('products', [])

    # Extract links from each product and create a DataFrame
    links = {'product_name': [product.get('product_name', 'N/A') for product in products],
    'link': [product.get('url', 'N/A') for product in products]}
        
    links_df = pd.DataFrame(links)
    return links_df
    
```

```python
# Links for each product
off_links = get_links(off_data)
off_links
```
