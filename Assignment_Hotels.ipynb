```python
import pandas as pd
import numpy as np
features = pd.read_csv("C:\\Users\\ceder\\Downloads\\hotels-europe_features.csv")
price = pd.read_csv("C:\\Users\\ceder\\Downloads\\hotels-europe_price.csv")
```

```python
#       Checking nature of data
features.dtypes
```

```python
price.dtypes
```

```python
#    Creating a new DataFrame "hotels" from merging "features" and "price" on "hotel_id" column (assuming hotel_id relates to the same data)
hotels = pd.merge(features, price, on = "hotel_id")
```

```python
#    Selecting hotels booked in Jan 2018 & on a weekend
hotel_sel = hotels[(hotels.year == 2018) & (hotels.month == 1) & (hotels.weekend == 1)]
```

```python
#     Selecting hotels in Vienna or London (from city column rather than city_actual)
hotel_sel = hotel_sel.loc[hotel_sel["city"].isin(["Vienna","London"])]
```

```python
#    Selecting hotels with 3 or 4 stars
hotel_sel = hotel_sel[(hotel_sel.stars == 3) | (hotel_sel.stars == 4)]
```

```python
#    checking all accommodations types (possible mispelling)
hotel_sel.groupby(by = ["accommodation_type"]).mean()
```

hotel_sel = hotel_sel[(hotel_sel.accommodation_type == "Apartment")]

```python
#    Selecting apartments with more than 10 reviews and price under $600 (assuming price is in $)
hotel_sel = hotel_sel[(hotel_sel.accommodation_type == "Apartment") & (hotel_sel.rating_reviewcount >= 10) & (hotel_sel.price < 600)]
```
