# Long to wide


```python
import pandas as pd
```

```python


# Create a sample DataFrame for long data
df = pd.DataFrame({
    'id': [1, 1, 2, 2, 3, 3],
    'variable': ['A', 'B', 'A', 'C', 'A', 'B'],
    'value': [10, 20, 30, 40, 50, 60]
})

print(df)
# Convert the long data to wide data
df.pivot(index='id', columns='variable', values='value')


```

```python


# Create a sample DataFrame for long data
df = pd.DataFrame({
    'id': [1, 1, 2, 2, 3, 3],
    'variable': ['A', 'B', 'A', 'C', 'A', 'B'],
    'value': [10, 20, 30, 40, 50, 60],
    'value2': [101, 202, 330, 440, 650, 60]
})

print(df)
# Convert the long data to wide data
df.pivot(index='id', columns='variable', values=['value', 'value2'] )


```

# Wide to long

```python
# create data for wide data
df = pd.DataFrame({
    'group': [1, 2, 3],
    'A': [10, 30, 50],
    'B': [20, 40, 60],
    'C': [30, 50, 70]
})

print(df)
# Convert the wide data to long data
df.melt(id_vars='group', value_vars=['A', 'B', 'C'], var_name='variable', value_name='value')
```

```python
# create data for wide data
df = pd.DataFrame({
    'group': [1, 2, 3],
    'A': [10, 30, 50],
    'B': [20, 40, 60],
    'C': [30, 50, 70],
    'D': [301, 510, 720]
})

print(df)
# Convert the wide data to long data
df.melt(id_vars='group', value_vars=['A', 'B', 'C'], var_name='variable', value_name='value')
```
