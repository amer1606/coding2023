```python
class FitnessTracker:
    def __init__(self, user):
        self.user = user
        self.steps = 0
        self.calories_burned = 0
        self.sleep_data = []
        self.sleep_average = 0

    def walk(self, new_steps):
        if new_steps >= 0:
            self.steps += new_steps
            self.calculate_calories_burned()
            return f"Steps updated to {new_steps}."

    def add_sleep_data(self, sleep_duration):
        if sleep_duration > 0:
            self.sleep_data.append(sleep_duration)
            self.calculate_avg_sleep()
            return f"Sleep data added: {sleep_duration} hours."

    def calculate_avg_sleep(self):
        self.sleep_average = sum(self.sleep_data)/len(self.sleep_data)

    def calculate_calories_burned(self):
        # Assuming a simple calculation: 0.05 calories burned per step
        self.calories_burned =  int(0.05 * self.steps)

    def display_fitness_info(self):
        return f"{self.user}'s Fitness Information:\n" \
               f"Steps: {self.steps}\n" \
               f"Calories Burned: {self.calories_burned} calories\n" \
               f"Average Sleep Duration: {self.sleep_average:.2f} hours\n" \
               f"Number of Days with Sleep Data: {len(self.sleep_data)}\n" \


# Example usage:
fitness_tracker = FitnessTracker("John Doe")

# Display initial fitness information
print(fitness_tracker.display_fitness_info())

# Update steps, add sleep data, and set fitness goal
fitness_tracker.walk(9000)
fitness_tracker.add_sleep_data(7)

fitness_tracker.walk(7500)
fitness_tracker.add_sleep_data(8)

# Display updated fitness information
print("\nAfter Updates:")
print(fitness_tracker.display_fitness_info())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 20]

# Matplotlib
plt.bar(categories, values, color='blue')
plt.title('Matplotlib Bar Chart')
plt.show()

# Seaborn
sns.barplot(x=categories, y=values, palette='viridis')
plt.title('Seaborn Bar Chart')
plt.show()

# Plotly
fig = px.bar(x=categories, y=values, title='Plotly Bar Chart')
fig.show()

# Bokeh
b_fig = figure(x_range=categories, height=350, title='Bokeh Bar Chart')
b_fig.vbar(x=categories, top=values, width=0.9, color='orange')
show(b_fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x_values = np.random.rand(20)
y_values = np.random.rand(20)

# Matplotlib
plt.scatter(x_values, y_values, color='red')
plt.title('Matplotlib Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Seaborn
sns.scatterplot(x=x_values, y=y_values, color='green')
plt.title('Seaborn Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Plotly
fig = px.scatter(x=x_values, y=y_values, title='Plotly Scatter Plot', labels={'x': 'X-axis', 'y': 'Y-axis'})
fig.show()

# Bokeh
b_fig = figure(title='Bokeh Scatter Plot', x_axis_label='X-axis', y_axis_label='Y-axis')
b_fig.scatter(x_values, y_values, size=8, color='blue')
show(b_fig)
```

```python
import pandas as pd
from ipyvizzu import Chart, Data, Config

df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.16/showcases/titanic/titanic.csv"
)
data = Data()
data.add_df(df)

chart = Chart(width="640px", height="360px")

chart.animate(data)

chart.animate(
    Config(
        {
            "x": "Count",
            "y": "Sex",
            "label": "Count",
            "title": "Passengers of the Titanic",
        }
    )
)
chart.animate(
    Config(
        {
            "x": ["Count", "Survived"],
            "label": ["Count", "Survived"],
            "color": "Survived",
        }
    )
)
chart.animate(Config({"x": "Count", "y": ["Sex", "Survived"]}))
```
