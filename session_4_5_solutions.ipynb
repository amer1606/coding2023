How many seats the plane with the highest average arrival delay to Miami (MIA) in January had?

```python

# Filter flights arriving at MIA in July
mia_july_flights = flights[(flights['dest'] == 'MIA') & (flights['month'] == 1)]

# Group by tail number and calculate the average arrival delay
average_delay_by_tailnum = mia_july_flights.groupby('tailnum')['arr_delay'].mean().reset_index()

# Find the tail number with the highest average arrival delay in July
tailnum_with_highest_delay = average_delay_by_tailnum.sort_values(by='arr_delay', ascending=False).head(1)
print(tailnum_with_highest_delay)
# Merge with planes
pd.merge(tailnum_with_highest_delay, planes, on='tailnum')['seats'][0]


```

Filter the flights departing from JFK Airport (origin = 'JFK') in January, calculate the average departure delay for each airline, and create a barchart  plot using Plotly and Plotnie to visualize the results.


```python

# Filter flights departing from JFK in January
jfk_jan_flights = flights[(flights['origin'] == 'JFK') & (flights['month'] == 1)]

# Calculate average departure delay by airline
avg_delay_by_airline = jfk_jan_flights.groupby('carrier')['dep_delay'].mean().reset_index().sort_values(by = 'dep_delay', ascending=False)



```

```python
# merge with airlines to see the names

avg_delay_by_airline_name = pd.merge(avg_delay_by_airline, airlines)
avg_delay_by_airline_name
```

```python
# Create a bar plot using Plotly
fig = px.bar(avg_delay_by_airline_name, x='name', y='dep_delay', title='Average Departure Delay by Airline (JFK - January)',
             labels={'dep_delay': 'Departure delay',
                     'name': 'Carrier'})
fig.show()
```

```python
ggplot(avg_delay_by_airline_name, aes(x= "reorder(name, dep_delay, ascending=False)", y='dep_delay')) + \
geom_col()+xlab('Carrier')+ylab('Departure delay')+ theme_void() +theme( axis_text_x=element_text(rotation=45, hjust=2))
```

Find the top 10 destinations from JFK Airport in August

```python

# Filter flights departing from JFK in August
jfk_august_flights = flights[(flights['origin'] == 'JFK') & (flights['month'] == 8)]

# Group by destination airport and count the number of flights
destination_counts = jfk_august_flights['dest'].value_counts().reset_index()
destination_counts.columns = ['Destination', 'Flights']

# Sort destinations by the number of flights in descending order
sorted_destinations = destination_counts.sort_values(by='Flights', ascending=False)


# Select the top 10 destinations
top_10_destinations = sorted_destinations.head(10)
```

```python
top_10_destinations
```

```python
fig = px.bar(top_10_destinations, x='Destination', y='Flights', title='Top 10 Destinations from JFK in August')
fig.show()
```

```python

ggplot(top_10_destinations, aes(x='reorder(Destination, Flights, ascending=False)', y='Flights')) + \
geom_col() + theme(axis_text_x=element_text(rotation=90)) + \
xlab('Destination') + labs(title='Top 10 Destinations from JFK in August')+theme_bw()
```

```python


# Combine 'year,' 'month,' and 'day' columns to create a new 'date' column
flights['date'] = pd.to_datetime(flights[['year', 'month', 'day']])

# Group the data by date and count the number of flights for each date
flights_by_date = flights.groupby('date').size().reset_index(name='Flights')

# Create a line chart using Plotly
fig = px.line(flights_by_date, x='date', y='Flights', title='Number of Flights by Date (Plotly)')
fig.show()

# Create a line chart using Plotnine
ggplot(flights_by_date, aes(x='date', y='Flights')) + geom_line() + labs(title='Number of Flights by Date (Plotnine)')
```
