```python
jfk = flights[(flights["origin"] == "JFK") & (flights["month"] == 1)]
```

```python
jfk_tail= jfk.groupby("tailnum").size().reset_index()
jfk_tail.columns = ["tailnum","count"]
jfk_tail.sort_values(by = "count", ascending = False)
```

```python
jfk["tailnum"].value_counts().reset_index()
```

```python
manuf = pd.merge(jfk_tail, planes, on = "tailnum")
manuf
```

```python
jfk_manuf = manuf.groupby("manufacturer").size()
```
